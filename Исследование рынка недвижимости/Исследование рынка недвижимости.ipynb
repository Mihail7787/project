# Предобработка данных

**Работа с пропусками во всего этажей в доме (floors_total)**`

```python
# заменяем пропуски в стоблце floors_total
# если медиана всего этажей меньше этажа квартиры, то меняем на этаж
# если медиана всего этажей больше этажа квартиры, то меняем на медиану


list_apartment.loc[(list_apartment['floors_total'].isna()) & (list_apartment['floors_total']\
    .median()< list_apartment['floor']),'floors_total'] = list_apartment['floor'] # заменяем пропуски в этажности на этаж
list_apartment.loc[(list_apartment['floors_total'].isna()) & (list_apartment['floors_total']\
    .median()>=list_apartment['floor']),'floors_total'] = list_apartment['floors_total'].median() #заменяем пропуски в этажности на медиану
```

**Работа с пропусками высота потолков (ceiling_height)**

```python
# заменяем пропуски в столбце ceiling_height
# заменяем проупуски с помощь lambda функции и группирвовки по этажности, исходим из того что для определенной этажности
# соответсвует определенная высота потолков, так например 3-4 этажные дама сталинского типа имеют высоту за 3 м
list_apartment['ceiling_height'] = list_apartment.groupby('floors_total')\
['ceiling_height'].apply(lambda x: x.fillna(x.median()))# заменил пропуски по высоте потолков на медиану
```

**Работа с пропусками в столбцах жилая площадь, площадь кухни (living_area, kitchen_area)**

```python
# прроуски в living_area, kitchen_area заменяем медианой долей площадей
# для этого находим доли жилой площади и кухни в общей площади
# затем находи медиану долей и заменяем ей пропуски
kof_living_area=round((list_apartment['living_area']/list_apartment['total_area']).median(),2) #рачитал коф. жилой площади
kof_kitchen_area=round((list_apartment['kitchen_area']/list_apartment['total_area']).median(),2) # расчитал коф.кухни
kof_kitchen_area
print('Доля жилой площади:',kof_living_area)
print('Доля кухонной площади:',kof_kitchen_area)
```

```python
list_apartment.loc[list_apartment['living_area'].isna(), 'living_area'] = (list_apartment['total_area']*kof_living_area).round(2)# заменил пропуски по жилой площади
list_apartment.loc[list_apartment['kitchen_area'].isna(), 'kitchen_area'] = (list_apartment['total_area']*kof_kitchen_area).round(2) # заменил пропуски по кухни
```

**Работа с пропусками в  столбцах аппартаменты, балкон (is_apartment, balcony)**

```python
# замену пропусков по аппартаменатм исходим из того, что владельцы у кторых аппартаменты указали этот факт
# остаьные же не стали замарачиваться по заполнению этого пункта. Поэтому пропуски смело можно заменить на False
list_apartment.loc[list_apartment['is_apartment'].isna(), 'is_apartment']= False #заменилb пропуски в аппартаментах
```

```python
# замену пропусков по балконом сделали на основании предположения, что если балкон не указан то его нет
list_apartment['balcony']= list_apartment['balcony'].fillna(0) # заменил пропуски в балконах
```

**работа с пропусками в столбце расстояние до центра города (cityCenters_nearest)**

```python
# заполним частично значения расстояния до центара
# для этого приведем населенный пункты к одному виду а именно слов посёлок
# заменим медианой пропуски, медиану расчитаеум по населленнм пунктам, пропуски для которых данные для заполнения не были выявлены
# трогать не будем, либо таких населенных пунктов единицы либо они так далеко рапположены от ценра что это становится не актульной инфой
list_apartment=list_apartment.replace(to_replace ='поселок', value = 'посёлок', regex = True)# привел слово поселок к одному виду
list_apartment['cityCenters_nearest'] = list_apartment.groupby('locality_name')['cityCenters_nearest'].apply(lambda x: x.fillna(x.median()))# заменил чать пропусков по медиане
```

```python
# меняем тип float на int, поменяли на основании того что балконы могут только целые числа
list_apartment['floors_total']=list_apartment['floors_total'].astype('int')
```

**Вывод**



Были проведены замены пропусков, в интересующих данных:
- `floors_total` замены были произведены медианой, с условиям этажа квартиры
- `ceiling_height` пропуски были заменены медианой согласно этажности здания, так как определенной высотности зданий соответсвует определенная высота потолков
- `living_area, kitchen_area` пропуски были заменены медианой доли этих площадей к общей площади
- `cityCenters_nearest` заменили медианой согласно расположения населленых пунктов
- `balcony,  is_apartment` пропуски заменили соглсно логики если есть пропуски то знаяить нет балкони и кваритира не является аппартаментами.
- остальные пропуски не стали трогать так как посчитал, что все логические замены были сделаны, данных достаточно для проведения исследования и форимирования выводов, а дальнейшие попыпки избавиться от пропусков могут больше принести вреда чем пользы.

```python
# добавляем столбец стомость кв/м
list_apartment['square_metr_cost']=(list_apartment['last_price'] / list_apartment['total_area']).round()
display(list_apartment.head(1))
```

```python
# создаём функуию которая возвращает категорию этажа(первый,последний, другой)
def floor(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'

def floor_id(floors_category):
    if floors_category == 'первый':
        return 1
    if floors_category == 'последний':
        return 3
    return 2
# создаем в датафрейме столбец с категорией этажей, и проверяем подстановку значений
list_apartment['floors_category'] = list_apartment.apply(floor, axis=1)
list_apartment['floors_id'] = list_apartment['floors_category'].apply(floor_id)
display(list_apartment[list_apartment['floors_category']=='последний'].head(1))
display(list_apartment[list_apartment['floors_category']=='первый'].head(1))
display(list_apartment[list_apartment['floors_category']=='другой'].head(1))
```

```python
# добавдяем в датафрейм стобцы год, месяц, день
list_apartment['year'] = pd.DatetimeIndex(list_apartment['first_day_exposition']).year
list_apartment['month'] = pd.DatetimeIndex(list_apartment['first_day_exposition']).month
list_apartment['day'] = pd.DatetimeIndex(list_apartment['first_day_exposition']).day
display(list_apartment.head(1))
```

```python
# добавляем в датафейм столбцы долю жилой площади и долю кухни
list_apartment['share_living_area'] = (list_apartment['living_area'] / list_apartment['total_area']).round(2)
list_apartment ['share_kitchen_area'] = (list_apartment['kitchen_area'] / list_apartment['total_area']).round(2)
display(list_apartment.head(1))
```

```python
# добавляем в датафрейм столбец растояние до центра в км
list_apartment['cityCenters_nearest_km']=(list_apartment['cityCenters_nearest'] / 1000).round()
display(list_apartment.head(1))
```

Были добавлены в таблицу следующие столбцы:

 - `floors_category, floors_id` (категория этажа, идентификатор категории)
 - `year, month, day` (год, месяц, день размещения объявления)
 - `share_living_area` (доля жилой площади)

```python
# Создаем фукуию, которая на вход приннимает датафрейм и его столбец
# функкия описывает данные по столбцу методом describe
# строить гистограмму в диапозоне оси x = q3-iqr*1.5),(q3+iqr*4) 

def hist_parametr(data, colunn):
    print(data[colunn].describe())
    q1 = data[colunn].quantile(0.25)
    q3 = data[colunn].quantile(0.75)
    iqr = q3-q1
    data[colunn].hist(bins=10, range=((q3-iqr*1.5),(q3+iqr*4)))                 
    plt.xlabel(colunn)
    plt.ylabel('Частота')
    plt.title(f'Распределение {colunn}')
    plt.grid(False)
    plt.show()
    sns.distplot(data[colunn], kde=True, bins=15)
    plt.title(f'Распределение {colunn}')
    plt.ylabel('Частота')
    plt.grid(False)
```

```python
#функция которая описывает и строит графики параметров двух разных выборок
def compare(data,data1, column):
    display(data[column].describe())
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3-q1
    sns.distplot(data[column], kde=True, bins=10,label='Санкт-Петербург весь город')
    plt.legend()
    plt.title(f'Распределение {column}')
    plt.grid(False)
    
    q1 = data1[column].quantile(0.25)
    q3 = data1[column].quantile(0.75)
    iqr = q3-q1
    plt.xlim(2,(q3+iqr*2))
    display(data1[column].describe())
    sns.distplot(data1[column], kde=True, bins=10, label='центральный район Санкт-Петербурга')
    plt.legend()
    plt.xlim(2,(q3+iqr*2))
    plt.grid(False)
```
